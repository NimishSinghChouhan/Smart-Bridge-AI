```python
# Build an ANN model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Initializing the seq model
reg_model = Sequential()
# Adding the input layer to the model
reg_model.add(Dense(4,activation='relu'))
# Adding the 1st hidden layer to the model
reg_model.add(Dense(64,activation='relu'))
# Adding the 2nd hidden layer to the model
reg_model.add(Dense(32,activation='relu'))
# Adding output layer to the model
reg_model.add(Dense(1,activation='linear'))
```

```python
# Compiling the skeleton gradient decent and loss
reg_model.compile(optimizer='adam', loss='mse')
```

```python
# Training the model
reg_model.fit(xtrain,ytrain,batch_size=4,epochs=10)
```

```python
# Testing the model

ypred=reg_model.predict(xtest)
```

```python
# Comparing predicted value with actual value
comp = pd.DataFrame(ytest)  # Creating a dataframe
comp.columns = ['Actual Value']  # Changing the column name
comp['Predicted Values'] = ypred  # Creating a column based on the prediction
comp
```

```python
# Splitting the data

x=df.iloc[:,1:5].values
x
```

```python
y= pd.get_dummies(df.iloc[:,5:]).values
```

```python
# Splitting the data into train and test
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
xtrain.shape,xtest.shape,ytest.shape, ytest.shape
```

```python
xtrain.shape,xtest.shape,yytrain.shape, yytest.shape
```

```python
yytrain = pd.DataFrame(ytrain)
yytest = pd.DataFrame(ytest)
```

```python
# ANN Model
# Skeleton of ANN Model
""" how many neurons we want for it we will see the number of columns
 here the number of columns are 4"""
model = Sequential()
model.add(Dense(8,input_dim=4,activation='relu')) # input layer
#model.add(Dense(32,activation='relu')) # hidden layer
model.add(Dense(26,activation='relu')) # hidden layer
model.add(Dense(3,activation='softmax')) # output layer
# if it is binary class classification use sigmoid function
# if it is a multi class classification use softmax function
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=10,batch_size=5,validation_data=(xtest,ytest))
```
